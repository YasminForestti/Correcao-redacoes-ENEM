```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
enem = pd.read_excel('../data/dados_enem.xlsx')
enem = enem.iloc[3:].rename(columns={
    'Unnamed: 2': '[0 - 100]', 
    'Unnamed: 3': '[100 - 200]',
    'Unnamed: 4': '[200 - 300]', 
    'Unnamed: 5': '[300 - 400]',
    'Unnamed: 6': '[400 - 500]',
    'Unnamed: 7': '[500 - 600]',
    'Unnamed: 8': '[600 - 700]',
    'Unnamed: 9': '[700 - 800]',
    'Unnamed: 10': '[800 - 900]',
    'Unnamed: 11': '[900 - 1000]',
    'Unnamed: 12': '[1000]'
})
enem['Ano'] = enem['Ano'].astype(int)
```

```python
data = enem.drop(columns=['Participantes do ENEM']).melt(id_vars=['Ano'])
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
for i, year in enumerate([2020, 2019]):
    g = sns.barplot(data=data[data['Ano'] ==year], x='variable', y='value', ax=axes[i])
    g.set_xticklabels(g.get_xticklabels(), rotation=30)
    axes[i].set(title=f'Distribuição de Notas da Redação em {year}', xlabel='Nota', ylabel='Número de estudantes')
```
